## Skewness
* Positive Skewness (Right Skewness) - The majority of the data points are concentrated on the left side, and there are a few data points with very high values on the right side.
    * mean is usually greater than the median
* Negative Skewness (Left Skewness) - Most of the data points are concentrated on the right side, and there are a few data points with very low values on the left side.
    * mean is usually less than the median

```python
print(df['SepalLengthCm'].skew())
print(df['PetalWidthCm'].skew())
```
