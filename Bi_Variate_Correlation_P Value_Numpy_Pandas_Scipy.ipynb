### Correlation Visualization - https://rpsychologist.com/correlation/

#### Correlation Coefficient (r): The correlation coefficient, often denoted as "r," ranges from -1 to 1. A positive value indicates a positive correlation (as one variable increases, the other tends to increase), a negative value indicates a negative correlation (as one variable increases, the other tends to decrease), and a value near 0 suggests a weak or no linear correlation.

#### P-value: In hypothesis testing, the p-value is a measure of evidence against a null hypothesis. For correlation testing, the null hypothesis typically states that there is no correlation between the two variables. The p-value associated with a correlation test quantifies the probability of observing a correlation as extreme as the one computed from the sample data if the null hypothesis were true.

#### Relationship:

If the p-value is small (typically less than a chosen significance level, e.g., 0.05), it suggests that you have enough evidence to reject the null hypothesis. In the context of correlation, this means that the observed correlation is statistically significant.

If the p-value is large, you may fail to reject the null hypothesis, suggesting that there is not enough evidence to conclude that there is a significant correlation between the variables.

In summary, a small p-value suggests that the observed correlation is unlikely to be due to random chance alone, supporting the idea that there is a significant relationship between the variables. On the other hand, a large p-value suggests that the observed correlation could plausibly occur by random chance, and there is insufficient evidence to claim a significant correlation.

### Linear Regression line in scatterplot

```python
from scipy import stats
print(stats.linregress(df.Age, df['Sleep duration']))
```

```python
# y = mx + b
# y = slope(x) + absintercept

m, b, r, p, strerr = stats.linregress(df.Age, df['Sleep duration'])

# x will be the range of age values that predicts y - sleep duration values
x = range(df.Age.min(), df.Age.max())
y = m * x + b

# plot a line 
plt.plot(x,y, color = 'black')
plt.scatter(df.Age, df['Sleep duration'])
plt.xlabel("Age")
plt.ylabel("Sleep duration")
# plt.legend()
plt.show()
```
